```python
virus_fasta = "/home/laura/projects/virus-watch-data/virus_ref/uniques_noduplicates.fa"
virus_index = "/home/laura/projects/virus-watch-data/virus_ref/kallisto_aa/noduplicates_human_dna_cdna.idx"
virus_t2g = "/home/laura/projects/virus-watch-data/virus_ref/nodup_clu_t2g.txt"

host_cdna = "/home/laura/projects/virus-watch-data/human_ref/Homo_sapiens.GRCh38.cdna.all.fa.gz"
host_dna = "/home/laura/projects/virus-watch-data/human_ref/Homo_sapiens.GRCh38.dna.primary_assembly.fa.gz"
host_cdna_dna = "/home/laura/projects/virus-watch-data/human_ref/Homo_sapiens.GRCh38.cdna_dna.fa.gz"

fastq_folder = "/home/laura/projects/virus-watch-data/alex_ismagilov/VSP"
# fastq_folder = "/home/laura/projects/virus-watch-data/alex_ismagilov/Metagenomic"

out_folder = f"{fastq_folder}/results_cdna_dna"

kallisto = "/home/laura/temp/kallisto/build/src/kallisto"
bustools = "/home/laura/projects/bustools/build/src/bustools"
```

```python
# Combine host genome and transcriptome
!cat $host_cdna $host_dna > $host_cdna_dna
```

Generate virus index d-listed against mouse:

```python
# Generate virus reference index
!/usr/bin/time -v $kallisto index \
    -t 12 \
    --aa \
    --d-list $host_cdna_dna \
    -i $virus_index \
    $virus_fasta
```

```python
import os
import glob
```

```python
fastqs = []
for filename in glob.glob(f"{fastq_folder}/*.fastq.gz"):
    fastqs.append(filename.split("/")[-1])
    
fastqs.sort()
fastqs
```

```python
len(fastqs)
```

```python
%%time

for fastq in fastqs:
    sample = fastq.split(".fastq.gz")[0]

    !mkdir -p $out_folder/$sample

    !$kallisto bus \
            -i $virus_index \
            -o $out_folder/$sample/ \
            --aa \
            -t 12 \
            -x bulk \
            $fastq_folder/$fastq \
            &> $out_folder/$sample/kb_out.txt

    !$bustools sort -o $out_folder/$sample/output_sorted.bus $out_folder/$sample/output.bus

    !$bustools count \
        --genecounts \
        --cm \
        -o $out_folder/$sample/bustools_count/ \
        -g $virus_t2g \
        -e $out_folder/$sample/matrix.ec \
        -t $out_folder/$sample/transcripts.txt \
        $out_folder/$sample/output_sorted.bus
```
